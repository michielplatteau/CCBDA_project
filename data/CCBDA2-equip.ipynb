```python
russia_equip['military auto'].fillna(0, inplace=True) # fill in missing values with 0
russia_equip['fuel tank'].fillna(0, inplace=True) # fill in missing values with 0
russia_equip['special equipment'].fillna(0, inplace=True) # fill in missing values with 0
russia_equip['mobile SRBM system'].fillna(0, inplace=True) # fill in missing values with 0
russia_equip['vehicles and fuel tanks'].fillna(0, inplace=True) # fill in missing values with 0
russia_equip['cruise missiles'].fillna(0, inplace=True) # fill in missing values with 0
```

```python
import plotly.express as px
fig=px.line(russia_equip,x='date',y=['drone','aircraft','helicopter'],template='ggplot2',title='<b>Russia Air Equipment losses')
fig.show()
names=['aircraft','helicopter','drone']
values=[russia_equip['aircraft'].max(),russia_equip['helicopter'].max(),russia_equip['drone'].max()]
fig=px.pie(names=names,values=values,hole=.7,template='ggplot2')
fig.update_layout(title_x=0.5,legend=dict(orientation='h',yanchor='bottom',y=1.02,xanchor='right',x=1))
```

```python
px.bar(russia_equip.groupby(['greatest losses direction'],as_index=False)[['tank']].count().sort_values(by='tank',ascending=False).reset_index(drop=True).head(15),x='greatest losses direction',y='tank',color='greatest losses direction',text='tank',template='seaborn',labels={'tank':'count'},title='<b>Cities which Reported More Losses ')
```

```python
#Data Manipulation, categorizing units based on air, ground, or naval units.
air_units = ['drone','aircraft','helicopter', 'cruise missiles']
naval_units = ['naval ship']
ground_units = ['APC','vehicles and fuel tanks', 'tank', 'field artillery', 'MRL',  'anti-aircraft warfare', 'special equipment']
air_and_naval_units = ['drone','aircraft','helicopter', 'cruise missiles', 'naval ship']

russia_equip['Total Air Units'] = russia_equip[air_units].sum(axis=1)
russia_equip['Total Naval Units'] = russia_equip[naval_units].sum(axis=1)
russia_equip['Total Ground Units'] = russia_equip[ground_units].sum(axis=1)
russia_equip.head()
```
